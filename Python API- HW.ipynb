Weather API - Observable Trends
1. We can see higher wind speeds ranging from 30 to 40 mph at higher latitudes only.
2. We can note that for every latitude range from -50 to 80, there is aleast one city with 100% humidity.
3. The closer we are to the equator (0) the higher the temperature and father away from the equator the temperatures 
   get lower.

```python
import pandas as pd
import numpy as np
import json
import requests as req
import matplotlib.pyplot as plt
import citipy
from citipy import citipy
from config import api_key
```

```python
output_data_file= "../Instructions/output_data/weather_cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

```python
url = "http://api.openweathermap.org/data/2.5/weather?"
units = "Imperial"
query_url = url+ "units=" +units + "&appid=" +api_key + "&q="

cities = []
weather = []
count = 1
```

```python
#for city in cities:
while count < 550:
    latitude = np.random.uniform(-90, 90, 1)
    longitude = np.random.uniform(-180, 180, 1)
    city = citipy.nearest_city(latitude[0],longitude[0]).city_name
    if city not in cities:
        response = req.get(query_url + city).json()
        if response["cod"] == 200:
            print("Record Processing" +str(count) +" of Set 1 | "+ city)
            count = count + 1
            print(query_url+ city)
            cities.append(city)
            weather.append(response)
            
print("------------")           
print("Data Retrieval Complete")
print("-------------")
```

```python
len(cities)
```

```python
latitude_data = [data.get("coord").get("lat") for data in weather]
longitude_data= [data.get("coord").get("lon") for data in weather]
data_city=[data.get("name") for data in weather]
data_temp= [data.get("main").get("temp_max") for data in weather]
data_speed_of_wind = [data.get("wind").get("speed") for data in weather]
data_cloud= [data.get("clouds").get("all") for data in weather]
data_country= [data.get("sys").get("country") for data in weather]
data_date= [data.get("dt") for data in weather]
data_humidity = [data.get("main").get("humidity") for data in weather]

weather_data={"Date": data_date,"City": data_city,"Cloudiness": data_cloud, "Country": data_country,"Humidity": data_humidity,"Lat":latitude_data,"Lng":longitude_data,"Max Temp": data_temp,"Wind Speed": data_speed_of_wind}

weather_info= pd.DataFrame(weather_data)
weather_info.count()
```

```python
weather_info.to_csv("weather_info.csv")
weather_info.head()
```

```python
#"Latitude vs. Temperature Plot"

plt.figure(num=None, figsize=(8,5), dpi=80)
plt.scatter(weather_info["Lat"], weather_info["Max Temp"],marker="o",edgecolors="black")
plt.title("City Latitude vs. Max Temperature (9/22/2018)")
plt.ylabel("Max Temperature(F)")
plt.ylim(-100, 150)
plt.xlabel("Latitude")
plt.xlim(-80, 100)
plt.grid(True)
plt.savefig("LatitudeVs.TemperaturePlot.png")
plt.show()
```
